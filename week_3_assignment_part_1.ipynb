```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df = df[0]
df
```

```python
# Only process the cells that have an assigned borough. Ignore cells with a 
# borough that is Not assigned.

df = df[df["Borough"] != "Not assigned"]
df
```

```python
# If a cell has a borough but a Not assigned neighborhood, then the neighborhood 
# will be the same as the borough. 

def gen_neighbourhood(x): 
    if x[2] == "Not assigned":
        x[2] = x[1]
    return x

df = df.apply(gen_neighbourhood, axis=1).sort_values("Postcode")
df
```

```python
# More than one neighborhood can exist in one postal code area. For example, in 
# the table on the Wikipedia page, you will notice that M5A is listed twice and 
# has two neighborhoods: Harbourfront and Regent Park. These two rows will be 
# combined into one row with the neighborhoods separated with a comma as shown 
# in row 11 in the above table.

def clean_df(df):
    i = 0
    postcodes = {}
    df_clean =  pd.DataFrame(columns=['Postcode', 'Borough', 'Neighbourhood'])
    for index, row in df.iterrows():
        if row[0] not in postcodes:
            postcodes[row[0]] = i
            i += 1
            df_clean = df_clean.append(row, ignore_index=True)
        else:
            df_clean.iloc[postcodes[row[0]]][2] = df_clean.iloc[postcodes[row[0]]][2] + ', ' + row[2]
    return df_clean

df_clean = clean_df(df)
df_clean
```
